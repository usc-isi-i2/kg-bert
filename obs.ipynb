```python
from collections import defaultdict
```

```python
with open("../data/RICA_material_KnowledgeTable .csv","r") as f:
    head = f.readline().split(",")
    lines=[]
    for line in f:
        lines.append(line.strip().split(","))
```

```python
lines[0]
```

```python
comparetype=set()
for line in lines:
    comparetype.add(line[3])
```

```python
comparetype
```

```python
material_type=set();
for line in lines:
    material_type.add(line[0])
    material_type.add(line[1])
```

```python
# build item name
item_name= set()
for material in material_type:
    item_name.add(material+"_"+"item")
```

```python
# build traget and ground truth

# only transparent
new_lines=[]

for item in item_name:
    #print(item)
    item_name_duplicate=set(list(item_name))
    item_name_duplicate.remove(item)

    for item2 in item_name_duplicate:
        #print(item, item2)
        for compare in comparetype:
            new_lines.append([item,item2, compare])
```

```python
with open("../data/material_knowledge/more_targets.txt","w") as f1,open("../data/material_knowledge/more_truth.txt","w") as f2:
    for line in new_lines:
        item1, item2, compare = line
        f1.write("\t".join([item1,item2,compare,str(0)])+"\n")
        f2.write("\t".join([item1,item2,compare,str(1)])+"\n")
```
